```python
# !pip install yfinance
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# import pandas_datareader as data

from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override()
# y_symbols = ['SCHAND.NS', 'TATAPOWER.NS', 'ITC.NS']
from datetime import datetime
startdate = datetime(2010,1,1)
enddate = datetime(2019,12,31)
df = pdr.get_data_yahoo('TATAPOWER.NS', start=startdate, end=enddate)
df.head()
```

```python
df.tail()
```

```python
df = df.reset_index()
df.head()
```

```python
df = df.drop(['Date', 'Adj Close'], axis = 1)
df.head()
```

```python
plt.plot(df.Close)   
# Graph showing closing price of the stock
```

```python
ma100 = df.Close.rolling(100).mean()
ma100
# 1st 100 values are null
```

```python
plt.figure(figsize = (12, 6))
plt.plot(df.Close)
plt.plot(ma100, 'r')
# Graph with 100 days ma(moving average)
```

```python
ma200 = df.Close.rolling(200).mean()
ma200
# 1st 200days values are null, from 201 we'll have value 
```

```python
plt.figure(figsize = (12, 6))
plt.plot(df.Close)
plt.plot(ma100, 'r')
plt.plot(ma200, 'g')

# Graph with 100 days ma and 200days(moving average)
```

```python
df.shape
```

```python
# splitting data into training and testing
data_training = pd.DataFrame(df['Close'][0:int(len(df)*0.70)])
data_testing = pd.DataFrame(df['Close'][int(len(df)*0.70): int(len(df))])
# 70%data training and 30%testing
print(data_training.shape)
print(data_testing.shape)
```

```python
data_training.head()
```

```python
data_testing.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0, 1))

# Values ranged between 0-1, to feed to model and train efficiently
```

```python
data_training_array = scaler.fit_transform(data_training)
data_training_array
```

```python
data_training_array.shape
```

```python
x_train = []
# 100ma 1st 100days are xtrain 
# same as for 200ma
y_train = []
# 101th day will be y_train\


for i in range(100, data_training_array.shape[0]):
    x_train.append(data_training_array[i-100: i])
    y_train.append(data_training_array[i, 0])
    
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
x_train.shape

# 100 values have become columns as they need to act as features for y_train
```

```python
# Making ML Model

from keras.layers import Dense, Dropout, LSTM
from keras.models import Sequential
```

```python
model = Sequential()
model.add(LSTM(units = 50, activation = 'relu', return_sequences = True,
               input_shape = (x_train.shape[1], 1)))
model.add(Dropout(0.2))

model.add(LSTM(units = 60, activation = 'relu', return_sequences = True))
model.add(Dropout(0.3))

model.add(LSTM(units = 80, activation = 'relu', return_sequences = True))
model.add(Dropout(0.4))

model.add(LSTM(units = 120, activation = 'relu'))
model.add(Dropout(0.5))
          
# Adding dense layer to add all the above layers
model.add(Dense(units = 1))
```

```python
model.summary()
```

```python
# Performing time series analysis, therefore the loss is kept as MSE
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
model.fit(x_train, y_train, epochs = 50)
```

```python
model.save('keras_model.h5')
```

```python
data_testing.head()
```

```python
past_100_days = data_training.tail(100)
```

```python
final_df = past_100_days._append(data_testing, ignore_index = True)
```

```python
final_df.head()
```

```python
# Applying scaling, MinMaxScaler
input_data = scaler.fit_transform(final_df)
input_data
```

```python
input_data.shape
# Gives 800 or something rows and 1 columns
```

```python
x_test = []
y_test = []

for i in range(100, input_data.shape[0]):
    x_test.append(input_data[i-100: i])
    y_test.append(input_data[i, 0])
```

```python
x_test, y_test = np.array(x_test), np.array(y_test)
print(x_test.shape)
print(y_test.shape)
```

```python
# Making Predictions

y_predicted = model.predict(x_test)
```

```python
y_predicted.shape
```

```python
y_test
```

```python
y_predicted
```

```python
# All above values are scaled down
# And we need to scale up those values
scaler.scale_
#above gives factor by which above values were scaled down 
```

```python
scale_factor = 1/0.02012072
y_predicted = y_predicted*scale_factor
y_test = y_test*scale_factor
```

```python
plt.figure(figsize = (12, 6))
plt.plot(y_test, 'b', label = 'Original Price')
plt.plot(y_predicted, 'r', label = 'Predicted Price')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```
